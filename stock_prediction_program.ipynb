```python
#데이터  수집
from prophet import Prophet
import pandas as pd
import requests

new_url='https://finance.naver.com/item/sise_day.naver?code=252670&page='

my_headers={'user-agent':'Mozilla/5.0'}

all_tables_year=pd.DataFrame() #데이터 축적용 데이터 프레임 생성

for page_number in range(1,100):

  full_url=new_url+str(page_number)

  print(f'{page_number}번째 페이지 읽어오기({full_url})')

  page = requests.get(full_url, headers=my_headers) #HTTP 요청 전송 후 응답 받아오기

  table = pd.read_html(page.text)[0] #테이블 추출

  print(f'전체 {len(all_tables_year.index)} 줄에 {len(table.index)} 줄 추가')

  all_tables_year = pd.concat([all_tables_year, table]) #데이터 축적용 데이터프레임에 테이블 추가
```

```python
#데이터 가공

all_tables_year.dropna(inplace=True) #결손치 제거

all_tables_year.rename(columns={'날짜':'ds'},inplace=True) #열 이름 영어로 변경

all_tables_year['y']=(all_tables_year['고가']+all_tables_year['저가'])/2

all_tables_year
```

```python
#최종 입력 데이터프레임 준비

df=pd.DataFrame()
df= all_tables_year[['ds','y']]
df.info
df
```
